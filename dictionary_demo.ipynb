### Data Structure: dict() : dictionary :
A dictionary is a collection which is ordered, mutable, indexed through keys and does not allow duplicates. 

In Python dictionaries are written with curly brackets, and they have key-value pairs.

Each key-value pair in a Dictionary is separated by a colon(:), whereas many key:value pairs are separated by a comma.

A Dictionary in Python works similar to the Dictionary in a real world. Keys of a Dictionary must be unique and immutable, but the values associated with keys can be repeated and be of any type.

Note 1: As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

Note 2: Dictionaries are sometimes found in other languages as associative memories or associative arrays.

```python
# Example 1
cardict = { "brand": "Ford",                                                                                                
            "model": "Mustang",
            "year": 1964, 
            "colors":["red", "blue", "yellow", "grey"]
          }
print(cardict)
```

```python
Knowledge = {'Frank': {'Python', 'Perl'}, 
             'Guido': {'Python'}, 
             'Monica': {'C', 'C++'}}
print(Knowledge)
```

##### Dictionary Items
Dictionary items are ordered, changeable, and does not allow duplicates.

Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

```python
# Print the "brand" value of the car:

cardict = { "brand": "Ford",                                                                                                
            "model": "Mustang",
            "year": 1964, 
            "colors":["red", "blue", "yellow", "grey"]
          }
print(cardict["model"])
```

##### A dictionary is a data structure with zero or more elements with the following attributes.
* Keys are unique(within one dictionary). Duplicate keys are not allowed.
* Keys are immutable  cannot be changed
* Keys are hashable - Key value pairs are stored at the hashed location in some way
* dictionary like the set are ordered collection
* dictionary is indexable based on the key  key could be a string, an integer or a tuple or any immutable type
* An empty dictionary is created by using {} or by the constructor dict()
* We can extract keys using the method dict.keys() and the values using the method dict.values(). Both these return iterable objects. There will be one-to-one mapping between the keys in the dict.keys() and the values in the dict.values().
* We can iterate through a dict  we actually iterate through the keys.

```python
d1={}
d2=dict()
print(d1,type(d1))
print(d2,type(d2))
```

Dictionaries are indexed by keys, which can be any immutable type; strings and numbers can always be keys.
Tuples can be used as keys if they contain only strings, numbers, or tuples; if a tuple contains any mutable object either directly or indirectly, it cannot be used as a key.

```python
# Duplicates Not Allowed
# Dictionaries cannot have two items with the same key:
# Duplicate values will overwrite existing values:
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
# Print the number of items in the dictionary:
print(len(thisdict))
```

##### Dictionary Methods
Python has a set of built-in methods that you can use on dictionaries.


```python
# values() method
 # Returns a list of all the values in the dictionary
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(thisdict.values())

# keys() method
 # Returns a list containing the dictionary's keys
print(thisdict.keys())
```

```python
# pop() method
# Removes the element with the specified key
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(thisdict.pop("electric"))
print(thisdict)
```

```python
# popitem() method
# Removes the last inserted key-value pair
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(thisdict.popitem())
print(thisdict)
print(thisdict.popitem())
print(thisdict)
```

```python
# get() method
#It is a conventional method to access a value for a key.
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]
}
print(thisdict.get("brand"))
```

###### items() Method
The items() method returns a view object. The view object contains the key-value pairs of the dictionary, as tuples in a list.

The view object will reflect any changes done to the dictionary, see example below.

```python
# When an item in the dictionary changes value, 
  # the view object also gets updated:

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.items()
car["year"] = 2018
print(x)
print(car)
```

##### setdefault() Method
The setdefault() method returns the value of the item with the specified key.

If the key does not exist, insert the key, with the specified value.

Syntax: 
dictionary.setdefault(keyname, value)

Parameter Values:


```python
# Get the value of the "model" item:

car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

x = car.setdefault("model", "Bronco")
print(x)
print(car)
```

```python
''' Get the value of the "color" item, if the "color" item does not exist,
    insert "color" with the value "white" '''
car = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = car.setdefault("color", "white")
print(x)
print(car)
```

##### Turn Lists into Dictionaries
The zip() function take iterables (can be zero or more), makes iterator that aggregates elements based on the iterables passed, and returns an iterator of tuples.

```python
dishes = ["pizza", "sauerkraut", "paella", "hamburger"]
countries = ["Italy", "Germany", "Spain", "USA"]
country_specialities_dict = dict(zip(countries, dishes))
print(country_specialities_dict)

lt=zip(countries, dishes)
print(lt)
print(list(lt))
```
